```r
library(dplyr)
library(tidyr)
library(ggplot2)
```

```r
options(repr.plot.width = 10, repr.plot.height = 6)
```

```r
df <- read.csv("data/cars.tidy.csv", stringsAsFactor = FALSE)
df$price_in_1000 = df$price / 1000
colnames(df)
```

```r
df <- filter(df, price < 1700000)
```
